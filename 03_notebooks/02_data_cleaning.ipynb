```python
import pandas as pd
import numpy as np


customers = pd.read_csv(
    "../01_ecommerce_ops_analytics/data/raw/customers.csv"
)


customers.head()
customers.info()
customers.isna().sum()
```

```python
customers.columns = (
    customers.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
customers["signup_date"] = pd.to_datetime(customers["signup_date"], errors="coerce")
customers["last_order_date"] = pd.to_datetime(customers["last_order_date"], errors="coerce")
```

```python
str_cols = customers.select_dtypes(include="object").columns
customers[str_cols] = customers[str_cols].apply(lambda col: col.str.strip())
```

```python
customers["total_orders"] = pd.to_numeric(customers["total_orders"], errors="coerce").astype("Int64")
customers["total_spent"]  = pd.to_numeric(customers["total_spent"], errors="coerce")
```

```python
customers["phone"] = customers["phone"].astype(str).str.replace(r"\D", "", regex=True)
```

```python
customers.to_csv(
    "../01_ecommerce_ops_analytics/data/processed/customers_cleaned.csv", 
    index=False
)
```
