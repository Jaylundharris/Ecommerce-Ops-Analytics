```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../01_ecommerce_ops_analytics/data/processed/customers.db")
```

```python
query = """
SELECT
    customer_id,
    first_name,
    last_name,
    total_orders,
    total_spent
FROM customers
ORDER BY total_spent DESC
LIMIT 5;
"""

top_customers = pd.read_sql(query, conn)
top_customers
```
